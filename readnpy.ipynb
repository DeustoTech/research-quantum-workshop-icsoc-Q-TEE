```python
#Hacemos diccionario para 
prob_dicc={"Real Amplitudes":results[0,:,0],
               "EfficientSu2":results[1,:,0]}
```

```python


def hacer_grafica(prob_dicc):
    
    plt.figure(figsize=(8,6))
    for key, values in prob_dicc.items():
        plt.hist(values,alpha=0.5,label=key)
    plt.title(f'Histogram of {len(values)} random parameter vectors').set_fontsize(14)
    #Tamaño de la fuente 14 
    plt.xlabel('Expected Values <E>').set_fontsize(14)
    plt.ylabel('Frequency').set_fontsize(14)
    plt.legend(fontsize=12)
    plt.show()
        
        
    
    
```

```python
hacer_grafica(prob_dicc)
```
